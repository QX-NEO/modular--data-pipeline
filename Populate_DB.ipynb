```python
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
import password
from sqlalchemy import create_engine
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select
```

```python
def set_header(df):
    new_header = df.iloc[0]
    df = df[1:]
    df.columns = new_header
    return df
    
```

```python
def alter_table(df):
    set_head = set_header(df)
    remove = set_head[:-1]
    return remove
```

```python
path_to_chromedriver = "C:/Users/neo qi xiang/Desktop/modular/chromedriver_win32/chromedriver.exe"
```

```python
url = "https://www.ccilindia.com/FPI_ARCV.aspx"
response = requests.get(url)
content = BeautifulSoup(response.text,'html.parser')
option = content.find_all("option")
option.pop(0)
list_option  = [i.next for i in option]
```

```python
driver = webdriver.Chrome(executable_path = path_to_chromedriver)
driver.get(url)
combine = "not establish as table"
for options in list_option:
    print(options)
    dropdown = Select(driver.find_element_by_id("drpArchival"))
    try:
        dropdown.select_by_visible_text(options)
        content =  BeautifulSoup(driver.page_source, 'lxml')
    except:
        print("error in {}".format(options))
    
    for i in content.find_all("a"):
        postback_tags = []
        string_tag = str(i)
        if bool(re.search('<a href="javascript:__doPostBack', string_tag)):
            postback_tags.append(i.next)
            
    current_table = content.find_all('table')
    dfs = pd.read_html(str(current_table))
    main =  alter_table(dfs[13])
    
    for pages in postback_tags:
        elem = driver.find_element_by_link_text(pages)
        elem.click()
        driver.implicitly_wait(1)
        soup = BeautifulSoup(driver.page_source, 'lxml')
        tables = soup.find_all('table')
        try:
            dfs = pd.read_html(str(tables))
            df = alter_table(dfs[13])
            main = main.append(df)
            main['date'] = options
            print("updated for {}".format(options))
        except:
            print("no table")
        
    try:
        combine = combine.append(main)
    except:
        combine = main
        print("first table")
        
    driver.implicitly_wait(2)

driver.close()
```

```python
data_collected = combine
```

```python
data_collected['date'] =  pd.to_datetime(data_collected['date'],format='%d-%b-%Y')
data_collected[["Indicative Value Of Aggregate Holding Of FPIS (INR CR#)", "Outstanding Position Of Govt# Securities (INR CR#)", "Sec Holdings (%)"]] = data_collected[["Indicative Value Of Aggregate Holding Of FPIS (INR CR#)", "Outstanding Position Of Govt# Securities (INR CR#)", "Sec Holdings (%)"]].astype(float)
data_collected = data_collected.drop_duplicates()
data_collected = data_collected.rename(columns={"Indicative Value Of Aggregate Holding Of FPIS (INR CR#)": "IVAH", "Outstanding Position Of Govt# Securities (INR CR#)": "OPGS", "Sec Holdings (%)": "sec_holding"}, errors="raise")
data_collected.to_csv("modular.csv",index =False)
```

```python
def connect_sql():
    POSTGRES_ADDRESS = password.getaddress()
    POSTGRES_PORT =  password.getport()
    POSTGRES_USERNAME = password.getusername()
    POSTGRES_PASSWORD = password.getpass()
    POSTGRES_DBNAME = password.getdbname()
    postgres_str = 'postgresql://{username}:{password}@{ipaddress}:{port}/{dbname}'.format(username=POSTGRES_USERNAME,
                                                                                            password=POSTGRES_PASSWORD, ipaddress=POSTGRES_ADDRESS, port=POSTGRES_PORT, dbname=POSTGRES_DBNAME)
    return postgres_str
```

```python
engine = create_engine(connect_sql())
data_collected.to_sql('modular', con=engine, if_exists='replace', index = False)
```
